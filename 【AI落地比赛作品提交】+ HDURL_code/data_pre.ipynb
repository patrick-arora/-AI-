```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("data/train data.xlsx")
df
```

```python
df.rename(columns={'室温（K）':'工厂温度','室温（K）.1':'机器温度'},inplace=True)
df
```

```python
df.info()
```

```python
#查找重复值
bh=df.duplicated(['机器编号'],keep=False)
sf=df.duplicated(['统一规范代码'],keep=False)
sf.dtype
```

```python
import collections
countp1=collections.Counter(bh)
countp2=collections.Counter(sf)
print(countp1,countp2)
```

```python
#量化处理
dengji={'L':1,'M':2,'H':3}
df['机器质量等级']=df['机器质量等级'].map(dengji)
liebie={'Normal':0,'TWF':1,'HDF':2,'PWF':3,'OSF':4,'RNF':5}
df['具体故障类别']=df['具体故障类别'].map(liebie)
df.head()
```

```python
#查找异常值
df0=df[df['是否发生故障']==0]
df0.loc[df0['具体故障类别'].isin([1,2,3,4,5])]
```
